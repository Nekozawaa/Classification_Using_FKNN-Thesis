```python
# import library
import numpy as np
import pandas as pd
```

```python
#import dan menampilkan isi data_fix1819.csv
df = pd.read_csv("csv/data_fix1819.csv")
df
```

```python
#mengambil data dari beberapa kolom mk wajib Intelligent System
df_wajib_is = df.filter(['LOGIKA MATEMATIKA',
                               'MATEMATIKA DISKRIT',
                               'ALGORITMA PEMROGRAMAN',
                               'ANALISIS KOMPLEKSITAS ALGORITMA',
                               'STRUKTUR DATA',
                               'PENGANTAR KECERDASAN BUATAN',
                               'TEORI BAHASA DAN AUTOMATA',
                               'PEMBELAJARAN MESIN',
                               'PEMROGRAMAN BERORIENTASI OBJEK'])
df_wajib_is
```

```python
#mengambil data dari beberapa kolom mk wajib Cyber Physical System
df_wajib_cps = df.filter(['SISTEM DIGITAL',
                                'ORGANISASI DAN ARSITEKTUR KOMPUTER',
                                'SISTEM OPERASI',
                                'JARINGAN KOMPUTER',
                                'SISTEM PARALEL DAN TERDISTRIBUSI'])
df_wajib_cps
```

```python
#mengambil data dari beberapa kolom mk wajib Software Engineering
df_wajib_se = df.filter(['PEMODELAN BASIS DATA',
                               'RPL: ANALISIS KEBUTUHAN',
                               'SISTEM BASIS DATA',
                               'INTERAKSI MANUSIA KOMPUTER'])
df_wajib_se
```

```python
#mengambil data dari beberapa kolom mk wajib Data Science
df_wajib_ds = df.filter(['KALKULUS',
                               'STATISTIKA',
                               'KALKULUS LANJUT',
                               'MATRIKS DAN RUANG VEKTOR',
                               'TEORI PELUANG'])
df_wajib_ds
```

```python
#membuat dataset baru yg hanya berisikan kolom mata kuliah dasar, wajib icm, wajib side, dan wajib tele
df_test = pd.concat([df_wajib_is, df_wajib_cps, df_wajib_se, df_wajib_ds], axis=1)
df_test
```

```python
#menyimpan dataset df_test ke dalam file data_test18.csv
df_test.to_csv('csv/data_test18.csv', index = False)
```
